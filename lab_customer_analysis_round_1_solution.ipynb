5. Correct the values in the column 'customer lifetime value'. 
They are given as a percent, so divide them by 100 and change dtype to numerical type. (only 2 of the files need this step.)

```python
file1['customer_lifetime_value'] = file1['customer_lifetime_value'].str.replace('%', '')
file1['customer_lifetime_value'] = file1['customer_lifetime_value'].astype(float)
file1['customer_lifetime_value'] = file1['customer_lifetime_value'] / 100

file2['customer_lifetime_value'] = file2['customer_lifetime_value'].str.replace('%', '')
file2['customer_lifetime_value'] = file2['customer_lifetime_value'].astype(float)
file2['customer_lifetime_value'] = file2['customer_lifetime_value'] / 100
```
